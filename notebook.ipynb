 Análisis del coportamiento de Lucha en Juegos Olímpicos desde 2000 hasta 2024

```python

import pandas as pd
from functions import remove_duplicados
import matplotlib.pyplot as plt

data = "games-2000-to-2024.json"
lucha = pd.read_json(data)

list_players = []
for a in lucha:
    for b in lucha[a]:
        for c in b:
            if len(c) > 1:
                list_players.append(c)

luchadores = remove_duplicados(list_players)
luchadores_hombres = []

libre_m = []
g_r_m = []
libre_f = []

for i in range(2000,2025,4):
    for x in lucha[i]["libre"]:
        if lucha[i]["libre"][x]["sexo"] == "M" and x != "decalificado":
            libre_m.append(x)
        if lucha[i]["libre"][x]["sexo"] == "F" and x != "decalificado":
            libre_f.append(x)
    for y in lucha[i]["greco-romana"]:
        if lucha[i]["greco-romana"][y]["sexo"] == "M" and x != "decalificado":
            g_r_m.append(y)
    
greco_m = remove_duplicados(g_r_m)
free_m = remove_duplicados(libre_m)
free_f = remove_duplicados(libre_f)

print(len(free_f))



```

```python
def resumen(lista: list, n: int):
    salida = []
    others= []
    for a in lista:
        if type(a) == str:
            for b in range(0,n):
                salida.append(lista[b])
            if n < len(lista) :
                salida.append("otros")
        if type(a) == int or type(a) == float:
            for c in range(0,n):
                salida.append(lista[c])
            for d in range(n,len(lista)):
                others.append(lista[d])
    suma = sum(remove_duplicados(others))
    if n < len(lista) and (type(salida[n]) == int or type(salida[n]) == float):
        salida.append(suma)   
    return remove_duplicados(salida)

print(resumen([15,5,18,29,18,50,98],4))
print(resumen(["a","b","c","d","e"],3))




```

```python
def medalla(p: str, style: str, país: list, oro: list, plata: list, bronce: list):
    count_oro = 0
    count_plata = 0
    count_bronce = 0
    pais = [] 
    for a in range(2000,2025,4):
        if p in lucha[a][style]:
            medalla = lucha[a][style][p]["medalla"]
            if medalla == "Oro":
                count_oro += 1
                pais.append(lucha[a][style][p]["país"])
            if medalla == "Plata":
                count_plata += 1
                pais.append(lucha[a][style][p]["país"])
            if medalla == "Bronce":
                count_bronce += 1
                pais.append(lucha[a][style][p]["país"])
    oro.append(count_oro)
    plata.append(count_plata)
    bronce.append(count_bronce)
    país.extend(remove_duplicados(pais))
    
oro_g = []
plata_g = []
bronce_g = []
pais_g = []

oro_l_m = []
plata_l_m = []
bronce_l_m = []
pais_l_m = []


oro_l_f = []
plata_l_f = []
bronce_l_f = []
pais_l_f = []

for a in greco_m:
    medalla(a,"greco-romana",pais_g, oro_g,plata_g,bronce_g)
for b in free_m:
    medalla(b, "libre", pais_l_m, oro_l_m, plata_l_m, bronce_l_m)
for c in free_f:
    medalla(b, "libre", pais_l_f, oro_l_f, plata_l_f, bronce_l_f)

```

```python
def best(p,oro,plata,bronce):
    table = {
    "Luchador": p,
    "Oro": oro,
    "Plata": plata,
    "Bronce": bronce
    }
    df = pd.DataFrame(table)
    pd.set_option('display.max_rows', None)
    sort_greco = df.sort_values(by="Oro",ignore_index=True, ascending=False)
    return sort_greco.head(20)
```
